**Intoduction**

This dataset 'housing' provides us housing data of california. It gives us the following details - 
1. longitude - A measure of how far west a house is.
2. Latitude - A measure of how far north a house is
3. Housing_median_age - Median age of a house. A lower number is a newer building.
4. total_rooms - total number of rooms.
5. total_bedrooms - total  no of bedrooms
6. population - total no of people residing the area
7. households - Total number of households, families residing within a home unit
8. median_income - Median income for households
9. median_house_value - Median house value for households 
10. ocean_proximity - The house has a Ocean view or not.

Here, Ocean_proximity is an ordinal value.

longitude, latitude, median_income and median_house_value are continuous values.

housing_median_age, total_rooms, total_bedrooms, population and households are discrete values.

#####Question 6 : Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
#function 
def cal_median( df ):

  col = list(df.columns)

  for c in col:
    l = df[c].dropna().sort_values()

    if len(l)//2 == 0:
      median = (l.iloc[len(l)//2] + l.iloc[(len(l)//2)-1])/2
    else:
      median = l.iloc[len(l)//2]

    print(c.ljust(20), '\t\t' , median)


#function call
cal_median(dfh)             #passing dataframe as argument
```

#####Question 10 : Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
def size_bedroom (total_bedrooms):
  if total_bedrooms <= 10 :
    return 'small'
  elif (total_bedrooms < 1000):
    return 'medium'
  else:
    return 'large'


total_bedroom_size = dfh['total_bedrooms'].map(size_bedroom)

dfh['total_bedroom_size'] = total_bedroom_size
dfh
```
